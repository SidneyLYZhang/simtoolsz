```python
def write_update_data(
    data: pl.DataFrame, update_data: pl.DataFrame, path: str,
    on: str = "ISO3", how: str = "left"
) -> pl.DataFrame :
    """
    å†™å…¥æ•°æ®åˆ°æŒ‡å®šè·¯å¾„ã€‚
    
    Args:
        data: è¦å†™å…¥çš„æ•°æ®
        path: å†™å…¥è·¯å¾„
    """
    # åˆå¹¶æ•°æ®
    data = data.join(update_data, on = on, how = how)
    data.write_parquet(path,compression="zstd",compression_level=6)
    return data
```

```python
import polars as pl
import re
from typing import Optional

def find_matching_rows(
    df: pl.DataFrame, 
    text: str, 
    regex_column: str = "regex",
    sort_by_quality: bool = False
) -> pl.DataFrame:
    """
    æ‰¾åˆ°æ‰€æœ‰ regex åˆ—èƒ½åŒ¹é…ç»™å®šæ–‡æœ¬çš„è¡Œ
    
    å‚æ•°:
        df: Polars DataFrame
        text: è¦åŒ¹é…çš„æ–‡æœ¬å­—ç¬¦ä¸²
        regex_column: åŒ…å«æ­£åˆ™è¡¨è¾¾å¼çš„åˆ—å
        sort_by_quality: æ˜¯å¦æŒ‰åŒ¹é…è´¨é‡æ’åºï¼ˆåŒ¹é…é•¿åº¦+æ•è·ç»„æ•°ï¼‰
        
    è¿”å›:
        åŒ¹é…çš„è¡Œç»„æˆçš„DataFrameï¼ˆåŒ…å«åŒ¹é…è¯¦æƒ…åˆ—ï¼‰
    """
    def check_pattern(pattern: str) -> tuple[bool, int, int, Optional[str]]:
        """æ£€æŸ¥å•ä¸ªæ­£åˆ™æ˜¯å¦åŒ¹é…ï¼Œè¿”å›(æ˜¯å¦åŒ¹é…, åŒ¹é…é•¿åº¦, æ•è·ç»„æ•°, é”™è¯¯ä¿¡æ¯)"""
        try:
            match = re.search(pattern, text)
            if match:
                return True, len(match.group(0)), len(match.groups()), None
            return False, 0, 0, None
        except re.error as e:
            return False, 0, 0, f"æ— æ•ˆçš„æ­£åˆ™è¡¨è¾¾å¼: {e}"
    
    # åº”ç”¨æ£€æŸ¥å‡½æ•°
    results = df[regex_column].map_batches(
        check_pattern,
        return_dtype=pl.Struct([
            pl.Field("matches", pl.Boolean),
            pl.Field("match_length", pl.UInt32),
            pl.Field("match_groups", pl.UInt32),
            pl.Field("error", pl.String),
        ])
    )
    
    # åˆ†è§£ç»“æœå¹¶æ·»åŠ åˆ°DataFrame
    df_with_results = df.with_columns(
        matches=results.struct.field("matches"),
        match_length=results.struct.field("match_length"),
        match_groups=results.struct.field("match_groups"),
        error=results.struct.field("error")
    )
    
    # è¿‡æ»¤å¹¶å¤„ç†
    matched = df_with_results.filter(pl.col("matches"))
    
    if sort_by_quality:
        # æŒ‰åŒ¹é…é•¿åº¦å’Œæ•è·ç»„æ•°é™åºæ’åºï¼ˆè¶ŠåŒ¹é…è¶Šé å‰ï¼‰
        matched = matched.sort(
            ["match_length", "match_groups"], 
            descending=[True, True]
        )
    
    return matched
```
